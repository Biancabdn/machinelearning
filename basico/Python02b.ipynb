```python
from datetime import date
hj = date.today()
print('Hoje é', hj)
hj45 = hj.toordinal() # Tranforma data para inteiro
print('Inteiro', hj45)
hj45 = hj45 + 45
print('Inteiro', hj45)
futuro = hj.fromordinal(hj45) # Transforma inteiro para data
print('Futuro', futuro)
```

## Estruturas de Repetição

### Laço Indeterminado
Não tenho a noção de qts vezes vai repetir

```python
temp = 0
while temp != -1000:
    temp = eval(input('Entre com a temperatura (-1000 para terminar):'))
    if temp != -1000:
        print('Em Fahrenheit:', 9/5*temp+32)
print('Até a próxima')        
```

```python
from random import randint
secreto = randint(1, 100)
chute = 0
while chute != secreto:
    chute = eval(input('Tente adivinhar o número:'))
    if chute > secreto:
        print('Tente algo menor')
    elif chute < secreto:
        print('Tente algo maior')

print('Parabéns vc acertou!')        
```

```python
palavra = 'FERNANDO'
print(palavra[-1])
print(palavra[:-1])
print(palavra[::-1])
```

```python
# NÚMEROS PALÍNDROMOS - 10000
quebra = 0
valMax = eval(input("Informe o valor"))
for i in range(1, valMax):
    s = str(i)
    if s == s[::-1]:
        print(s, end=' ')
        quebra += 1
        if quebra == 20:
            print()
            quebra = 0
```

```python
quebra = 0
i = 1
while i < 1000:
    s = str(i)
    if s == s[::-1]:
        print(s, end=' ')
        quebra += 1
        if quebra == 20:
            print()
            quebra = 0
    i += 1
```

```python
# Quantos números são terminados em 4 entre 0 e 10000?
ct = 0
for i in range(4, 10000, 10):
    s = str(i)
    if s[-1] == '4':
        ct += 1
print('Existem', ct, ' números terminados com 4')   

# Parente HTML - JavaScript
# Performático > 5 dias <> 2 hrs
```

### Comandos de Continuidade ou Interrupção

```python
x = 2 # solicitar
if x > 2:
    pass
else:    
    print("Olá Mundo")
```

```python
i = 2
while i < 22:
    i += 1
    if i % 2 == 0:
        continue  # Retorna para o WHILE
        
    print(i) # Mil comandos
    
    if i == 33:
        break    # Quebra o laço WHILE
else:
    print('Acabei...')
    
print('Terminei tudo')
```

```python
valor = 0
while valor > -1:
    valor = eval(input("Entre com um valor ou -1 para sair:"))
    if valor == 0:
        print('ZERO não VALE')
        break
    
    if valor == -1:
        continue     # PROXIMA INTERACAO
    else:
        pass
        
    print(10 / valor) # Um milhão de comandos

else:   
    print('Terminei.')    
```
